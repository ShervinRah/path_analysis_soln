```python
import pandas as pd
from matplotlib import pyplot as plt1
from matplotlib import pyplot as plt2
from matplotlib import cm 
import numpy as np
from mpl_toolkits.mplot3d.axes3d import get_test_data
```

```python
path = pd.read_csv("path.csv")
```

```python
path
```

```python
x1 = path.x_pth
y1 = path.y_pth
z1 = path.z_pth
```

```python
plt1.rcParams['figure.figsize'] = (8,6)
```

Importing the two data files provided, we first look at our path.csv file. Intially, I was interested to see the motion the robot was commanded to go to see what the outcome is supposed to look like. The robot seemed to be commanded certain 3D positions (x,y,z) to a draw a sort of "figure eight" motion while gradually increasing the size of the "figure eight". Here I plotted the path of the robotic arm. 

```python
ax1 = plt1.axes(projection='3d')
ax1.scatter3D(x1, y1, z1);
ax1.set_title("Path.csv")
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z');
```

Now that we have a better understanding of what the 3D motion is supposed to look like we are going to break this graph down to compare the path of each axis along with the associated time stamp. I plotted each each plane with its time. 

```python
fig1, axs1 = plt.subplots(3, 1)
axs1[0].plot(path.t_pth,x1, 'tab:purple')
axs1[0].set_title('X_pth vs T_pth')
axs1[0].set_xlabel('Time')
axs1[0].set_ylabel('X Path')
axs1[1].plot(path.t_pth,y1,'tab:orange')
axs1[1].set_title('Y_pth vs T_pth')
axs1[1].set_xlabel('Time')
axs1[1].set_ylabel('Y Path')
axs1[2].plot(path.t_pth,z1, 'tab:green')
axs1[2].set_title('Z_pth vs T_pth')
axs1[2].set_xlabel('Time')
axs1[2].set_ylabel('Z Path')
fig1.tight_layout()
```

From our x_pth,y_pth we see robot is drawing a wide figure eight motion and as time passes it begins to get smaller and smaller but our y_pth stays constant and does not change during the time of sending our path coordinates to the robot besides the initial z point. Now that we know what our robot look like along with the motion its suppose to draw for each plane let us take a look on what out what opticial tracker picked up to see if these graphs match up and look at any potential discrepancies between graphs and get to the root of the causes of why the tracker isn't tracking accordingly to the path that was sent to the robot. 

```python
tracker = pd.read_csv("tracker.csv")
```

```python
x2 = tracker.x_trk
y2 = tracker.y_trk
z2 = tracker.z_trk
```

```python
plt2.rcParams['figure.figsize'] = (8,6)
```

```python
ax2 = plt2.axes(projection='3d')
ax2.scatter3D(x2, y2, z2);
ax2.set_title("Tracker.csv")
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
```

```python
fig2, axs2 = plt.subplots(3, 1)
axs2[0].plot(tracker.t_trk,x2, 'tab:purple')
axs2[0].set_title('X_trk vs T_trk')
axs2[0].set_xlabel('Time')
axs2[0].set_ylabel('X Path')
axs2[1].plot(tracker.t_trk,y2,'tab:orange')
axs2[1].set_title('Y_trk vs T_trk')
axs2[1].set_xlabel('Time')
axs2[1].set_ylabel('Y Path')
axs2[2].plot(tracker.t_trk,z2,'tab:green')
axs2[2].set_title('Z_trk vs T_trk')
axs2[2].set_xlabel('Time')
axs2[2].set_ylabel('Z Path')
fig1.tight_layout()
```

Now that we have a graph of our tracker points on what the optical tracker picked up, we see that there is a bit of an issue. The coordinates that the optical tracker recorded do no match up to the one that was sent to the robot, and the motion of the circles(or figure eight) is at an angle which it should not be. My initial thought was there can be a faulty motor on one of the links for the robot causing this motion to be slanted. To better see if that was the case I graphed each plane along with the time stamps to see what happened. Luckily, we see that there is no drift in the motion or inconsistencies in the path that was sent meaning there is probably no faulty motors in one of the joints of the robot. There are two different possiblities on why this is happening. We know the the way the robot is drawing isnt the issue but it's position that was picked up. The first reason is that the origin of the robot could have been off, the initial position of the robot was offset and wasn't in zero configuration. Since an offset can affect the accuracy of the robot this could have been a cause to way it tracker wasn't tracking properly.This could have caused the z plane points to have dropped in the negative values which could have caused the coordinates to be thrown off when the path coordinates was being sent to the robot which may have caused this sort of slant. The second idea that comes to my head is that the robot and optical tracker aren't in the same frame. The optical tracker can be some sort of spheroically mounted retroreflector that isn't calibrated accordingly with the position of the robot. We need to find a way to better handle the tracker and robot coordinate frames to prevent this from happening. There might be a need to resonstruct the reference frame of the robot or the tracker frame of the optical tracker. The rotation matrix from the tracker frame to the robot frame might need to be adjusted.If the robot is in its correct zero configuration, then we know that the optical tracker may be causing this issue. To conclude, the robot and its motors are working correctly due to maintain the same drawing, however the position is thrown off by the possibilites mentioned above.
